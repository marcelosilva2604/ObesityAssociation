```python
import pandas as pd
import numpy as np

# Load the CSV file
df = pd.read_csv('/Users/marcelosilva/Desktop/early-obesity-prediction/A-Age Sample Filter/data_age_filtered_2_3_4.csv')

# Replace whitespace-only strings with NaN
df = df.replace(r'^\s*$', np.nan, regex=True)

# Calculate missing value percentage for each column
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Identify columns with more than 30% missing values
columns_to_drop = missing_percentage[missing_percentage > 30].index.tolist()

print(f"Columns with more than 30% missing values: {columns_to_drop}")
print(f"Number of columns to be dropped: {len(columns_to_drop)}")

# Drop columns with more than 30% missing values
df_cleaned = df.drop(columns=columns_to_drop)

print(f"Original shape: {df.shape}")
print(f"Cleaned shape: {df_cleaned.shape}")

# Display missing value summary for remaining columns
print("\nMissing value summary for remaining columns:")
remaining_missing = (df_cleaned.isnull().sum() / len(df_cleaned)) * 100
print(remaining_missing[remaining_missing > 0].sort_values(ascending=False))

# Save cleaned dataset
df_cleaned.to_csv('1missing.csv', index=False)

# Create summary text file
with open('1missing.txt', 'w') as f:
    f.write(f"Data Processing Summary\n")
    f.write(f"======================\n\n")
    f.write(f"Original number of columns: {df.shape[1]}\n")
    f.write(f"Final number of columns: {df_cleaned.shape[1]}\n")
    f.write(f"Columns removed: {len(columns_to_drop)}\n\n")
    f.write(f"Columns that were excluded (>30% missing values):\n")
    for i, col in enumerate(columns_to_drop, 1):
        f.write(f"{i}. {col}\n")

print(f"\nFiles saved:")
print(f"- Cleaned dataset: 1missing.csv")
print(f"- Summary report: 1missing.txt")
```

# Data Cleaning: Handling Missing Values

This section performs initial data cleaning by identifying and removing columns with excessive missing values.

## Process Overview

1. **Load the dataset** from the CSV file
2. **Clean whitespace** by replacing whitespace-only strings with NaN values
3. **Calculate missing value percentages** for each column
4. **Identify problematic columns** with more than 30% missing values
5. **Remove high-missing columns** to improve data quality
6. **Generate summary reports** for documentation

## Output Files

- `1missing.csv`: Cleaned dataset with problematic columns removed
- `1missing.txt`: Detailed summary report of the cleaning process

## Key Metrics

The process will display:
- Number of columns removed
- Original vs. cleaned dataset dimensions
- Missing value summary for remaining columns

This step ensures we work with a cleaner dataset for subsequent analysis while maintaining data integrity.
