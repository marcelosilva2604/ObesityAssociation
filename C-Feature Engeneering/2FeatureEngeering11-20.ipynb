```python
import pandas as pd
import numpy as np

# Load the dataset with feature engineering already applied
input_path = '/Users/marcelosilva/Desktop/early-obesity-prediction/C-Feature Engeneering/1FE1-10.csv'
df = pd.read_csv(input_path)

print("Dataset loaded successfully!")
print(f"Shape: {df.shape}")
print("\n" + "="*70)

# List of the next 10 variables for analysis
next_variables = [
    'h10b_sindrome_de_down',
    'h10b1_sindrome_nao', 
    'h10b2_sindrome_down',
    'h10b3_sindrome_fibrose',
    'h10b4_sindrome_fenilcetonuria',
    'h10b5_sindrome_autismo',
    'h10b9_sindrome_nao_sabe',
    'j08_ler',
    'j09_frequenta', 
    'j10_serie'
]

def analyze_variable_detailed(df, var_name, max_unique=15):
    """
    Detailed analysis of a specific variable
    
    Args:
        df: DataFrame
        var_name: variable name
        max_unique: maximum number of unique values to display
    """
    print(f"\n DETAILED ANALYSIS: {var_name}")
    print("-" * 60)
    
    # Check if the variable exists
    if var_name not in df.columns:
        print(f"VARIABLE NOT FOUND IN DATASET")
        return
    
    # Basic information
    total_rows = len(df)
    unique_count = df[var_name].nunique()
    missing_count = df[var_name].isna().sum()
    missing_pct = (missing_count / total_rows) * 100
    
    print(f"Data type: {df[var_name].dtype}")
    print(f"Total records: {total_rows:,}")
    print(f"Unique values: {unique_count}")
    print(f"Missing values: {missing_count} ({missing_pct:.2f}%)")
    print(f"Valid values: {total_rows - missing_count:,} ({100-missing_pct:.2f}%)")
    
    # Type-specific analysis
    if df[var_name].dtype in ['int64', 'float64']:
        # Numeric variable
        valid_data = df[var_name].dropna()
        if len(valid_data) > 0:
            print(f"\nDESCRIPTIVE STATISTICS:")
            stats = valid_data.describe()
            print(f"Mean: {stats['mean']:.2f}")
            print(f"Median: {stats['50%']:.2f}")
            print(f"Standard deviation: {stats['std']:.2f}")
            print(f"Minimum: {stats['min']}")
            print(f"Maximum: {stats['max']}")
            print(f"Q1: {stats['25%']}")
            print(f"Q3: {stats['75%']}")
        
        # If there are few unique values, show distribution
        if unique_count <= max_unique:
            print(f"\nVALUE DISTRIBUTION:")
            value_counts = df[var_name].value_counts().sort_index()
            for value, count in value_counts.items():
                percentage = (count / total_rows) * 100
                print(f"  {value}: {count:,} ({percentage:.2f}%)")
    
    else:
        # Categorical variable
        print(f"\nVALUE DISTRIBUTION:")
        if unique_count <= max_unique:
            # Show all categories
            value_counts = df[var_name].value_counts(dropna=False)
            for value, count in value_counts.items():
                percentage = (count / total_rows) * 100
                if pd.isna(value):
                    print(f"  [MISSING]: {count:,} ({percentage:.2f}%)")
                else:
                    print(f"  '{value}': {count:,} ({percentage:.2f}%)")
        else:
            # Show only the most frequent
            print(f"Top {max_unique} most frequent values:")
            value_counts = df[var_name].value_counts().head(max_unique)
            for value, count in value_counts.items():
                percentage = (count / total_rows) * 100
                print(f"  '{value}': {count:,} ({percentage:.2f}%)")
            
            if missing_count > 0:
                print(f"  [MISSING]: {missing_count:,} ({missing_pct:.2f}%)")
    
    # Identify potential issues
    print(f"\nOBSERVATIONS:")
    
    if missing_count > 0:
        if missing_pct > 10:
            print(f"  Warning: High percentage of missing values ({missing_pct:.1f}%)")
        elif missing_pct > 5:
            print(f"  Warning: Moderate percentage of missing values ({missing_pct:.1f}%)")
        else:
            print(f"  Info: Low percentage of missing values ({missing_pct:.1f}%)")
    else:
        print(f"  OK: No missing values")
    
    if unique_count == 1:
        print(f"  Warning: Constant variable (only 1 unique value)")
    elif unique_count == 2:
        print(f"  Info: Binary variable")
    elif unique_count < 10:
        print(f"  Info: Categorical variable with {unique_count} categories")
    
    # For categorical variables, identify rare categories
    if df[var_name].dtype == 'object' and unique_count > 2:
        value_counts = df[var_name].value_counts()
        rare_categories = value_counts[value_counts / total_rows < 0.01]  # <1%
        if len(rare_categories) > 0:
            print(f"  Warning: {len(rare_categories)} category(ies) with <1% of data:")
            for cat, count in rare_categories.items():
                pct = (count / total_rows) * 100
                print(f"      '{cat}': {count} ({pct:.2f}%)")

# Analyze each variable
print("ANALYSIS OF THE NEXT 10 VARIABLES FOR FEATURE ENGINEERING")
print("=" * 70)

for i, var in enumerate(next_variables, 1):
    print(f"\n[{i}/10] ", end="")
    analyze_variable_detailed(df, var)

print("\n" + "=" * 70)
print("ANALYSIS COMPLETED")
print("=" * 70)
print("\nSUMMARY FOR PLANNING:")
print("1. Review the distributions above")
print("2. Identify patterns in syndrome variables (h10b_*)")
print("3. Plan transformations for educational variables (j08_*, j09_*, j10_*)")
print("4. Define strategies for handling missing values")
print("5. Consider necessary groupings or transformations")
```

## Transformation Strategy

### Simplification and Position Preservation

- Remove redundant and highly correlated variables
- Replace original variables in their exact positions
- Create meaningful binary and ordinal variables

- `h10b5_sindrome_autismo`: 98.75% "No" (103 cases)
- `h10b9_sindrome_nao_sabe`: 99.76% "No" (20 cases)

#### Transformation Applied
- **Removed ALL 7 syndrome variables**
- **Created**: `sindrome_nao` (binary: 0=has syndrome, 1=no syndrome)
- **Position**: Inserted at original `h10b1_sindrome_nao` position
- **Distribution**: 98.31% no syndrome, 1.69% has syndrome
- **Rationale**: Extremely rare individual conditions make separate analysis impractical for ML

### 2. Maternal Education Variables (3 → 3 variables)

```python
import pandas as pd
import numpy as np
import os

def feature_engineering_vars_10_20():
    """
    Feature Engineering of variables 10-20 from the dataset
    """
    
    # Load dataset
    input_path = '/Users/marcelosilva/Desktop/early-obesity-prediction/C-Feature Engeneering/1FE1-10.csv'
    df = pd.read_csv(input_path)
    
    print(f"Original dataset: {df.shape}")
    print("Starting Feature Engineering of variables 10-20...")
    print("=" * 60)
    
    # Create copy for transformations
    df_fe = df.copy()
    
    # Variables to process
    syndrome_vars = [
        'h10b_sindrome_de_down', 'h10b1_sindrome_nao', 'h10b2_sindrome_down',
        'h10b3_sindrome_fibrose', 'h10b4_sindrome_fenilcetonuria', 
        'h10b5_sindrome_autismo', 'h10b9_sindrome_nao_sabe'
    ]
    
    education_vars = ['j08_ler', 'j09_frequenta', 'j10_serie']
    
    # Get original column positions to maintain order
    original_cols = df_fe.columns.tolist()
    
    # 1. Transform h10b1_sindrome_nao to binary (replace in same position)
    h10b1_pos = original_cols.index('h10b1_sindrome_nao')
    sindrome_nao_values = (df['h10b1_sindrome_nao'] == 'Sim').astype(int)
    
    # 2. Remove all syndrome variables first
    vars_to_remove = syndrome_vars.copy()  # Remove all 7 variables
    df_fe = df_fe.drop(columns=vars_to_remove)
    print(f"✓ Removed {len(vars_to_remove)} syndrome variables:")
    for var in vars_to_remove:
        print(f"   - {var}")
    
    # Insert sindrome_nao at the position where h10b1_sindrome_nao was
    current_cols = df_fe.columns.tolist()
    df_fe.insert(h10b1_pos, 'sindrome_nao', sindrome_nao_values)
    print("✓ h10b1_sindrome_nao: converted to sindrome_nao (0=has syndrome, 1=no syndrome)")
    
    # 3. Transform j08_ler to binary (replace in same position)
    j08_pos = df_fe.columns.get_loc('j08_ler')
    sabe_ler_values = (df['j08_ler'] == 'Sim').astype(int)
    df_fe = df_fe.drop(columns=['j08_ler'])
    df_fe.insert(j08_pos, 'sabe_ler', sabe_ler_values)
    print("✓ j08_ler: converted to sabe_ler (0=cannot read, 1=can read)")
    
    # 4. Transform j09_frequenta to binary (replace in same position)
    j09_pos = df_fe.columns.get_loc('j09_frequenta')
    teve_exposicao = df['j09_frequenta'].isin(['Frequenta', 'Já frequentou'])
    teve_exposicao_values = teve_exposicao.astype(int)
    df_fe = df_fe.drop(columns=['j09_frequenta'])
    df_fe.insert(j09_pos, 'teve_exposicao_escola', teve_exposicao_values)
    print("✓ j09_frequenta: converted to teve_exposicao_escola (0=never attended, 1=attended/attending)")
    print(f"   School exposure: {teve_exposicao.sum():,} ({teve_exposicao.mean()*100:.1f}%)")
    
    # 5. Transform j10_serie to ordinal educational level (replace in same position)
    def map_education_level(serie):
        """Map education series to ordinal levels"""
        if pd.isna(serie):
            return np.nan
        
        serie_lower = serie.lower().strip()
        
        # No study
        if 'sem estudo' in serie_lower:
            return 0
        
        # Elementary (fundamental)
        elementary_indicators = [
            '1° ano do ensino fundamental', '1ª série', '2°ano do ensino fundamental',
            '2ª série', '3ª série', '4°ano do ensino fundamental', '4ª série',
            '5°ano do ensino fundamental', '5ª série', '6°ano do ensino fundamental', 
            '6ª série', '7°ano do ensino fundamental', '7ª série', 
            '8°ano do ensino fundamental', '8ª série', '9°ano do ensino fundamental'
        ]
        
        if any(indicator in serie_lower for indicator in elementary_indicators):
            return 1
        
        # High school (médio)
        high_school_indicators = ['ensino médio', 'médio']
        if any(indicator in serie_lower for indicator in high_school_indicators):
            return 2
        
        # Higher education (superior)
        higher_ed_indicators = ['ensino superior', 'superior']
        if any(indicator in serie_lower for indicator in higher_ed_indicators):
            return 3
        
        return np.nan
    
    j10_pos = df_fe.columns.get_loc('j10_serie')
    nivel_educacional_values = df['j10_serie'].apply(map_education_level)
    df_fe = df_fe.drop(columns=['j10_serie'])
    df_fe.insert(j10_pos, 'nivel_educacional_mae', nivel_educacional_values)
    
    # Check mapping results
    education_distribution = df_fe['nivel_educacional_mae'].value_counts().sort_index()
    print("✓ j10_serie: converted to nivel_educacional_mae (ordinal)")
    print("   Education level distribution:")
    level_names = {0: 'No study', 1: 'Elementary', 2: 'High school', 3: 'Higher education'}
    for level, count in education_distribution.items():
        pct = (count / len(df_fe)) * 100
        print(f"   {int(level)} ({level_names[level]}): {count:,} ({pct:.1f}%)")
    
    # Handle any unmapped values
    unmapped = df_fe['nivel_educacional_mae'].isna().sum()
    if unmapped > 0:
        print(f"   ⚠️ Unmapped values: {unmapped}")
        # Show unmapped categories
        unmapped_categories = df[df_fe['nivel_educacional_mae'].isna()]['j10_serie'].unique()
        print(f"   Unmapped categories: {list(unmapped_categories)}")
    
    # Save result
    output_dir = '/Users/marcelosilva/Desktop/early-obesity-prediction/C-Feature Engeneering'
    output_path = os.path.join(output_dir, '2FE10-20.csv')
    df_fe.to_csv(output_path, index=False)
    
    print("\n" + "=" * 60)
    print("FEATURE ENGINEERING COMPLETED!")
    print("=" * 60)
    print(f"Original dataset: {df.shape}")
    print(f"Final dataset: {df_fe.shape}")
    print(f"File saved: {output_path}")
    print(f"Variables processed: 10")
    print(f"Variables removed: {len(vars_to_remove) + 3}")
    print(f"Variables created/transformed: 4")
    
    # Summary of changes
    print("\nSummary of transformations:")
    print("1. h10b1_sindrome_nao → sindrome_nao (binary)")
    print("2. j08_ler → sabe_ler (binary)")
    print("3. j09_frequenta → teve_exposicao_escola (binary)")
    print("4. j10_serie → nivel_educacional_mae (ordinal 0-3)")
    print(f"5. Removed {len(vars_to_remove)} redundant syndrome variables")
    
    # Final verification
    print("\n" + "=" * 60)
    print("FINAL VERIFICATION:")
    print("=" * 60)
    
    # Missing values
    missing_summary = df_fe.isnull().sum()
    total_missing = missing_summary.sum()
    if total_missing > 0:
        print("Missing values found:")
        for col, missing in missing_summary[missing_summary > 0].items():
            print(f"  {col}: {missing} ({missing/len(df_fe)*100:.2f}%)")
    else:
        print("✓ No missing values in final dataset")
    
    # Data types
    print(f"\nData types distribution:")
    dtype_counts = df_fe.dtypes.value_counts()
    for dtype, count in dtype_counts.items():
        print(f"  {dtype}: {count} variables")
    
    return df_fe

# Execute feature engineering
if __name__ == "__main__":
    df_result = feature_engineering_vars_10_20()
```
