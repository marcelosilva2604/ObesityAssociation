```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split

# Definir caminhos
caminho_origem = "/Users/marcelosilva/Desktop/projectOne/6/A-Binary Target/DataSetTarget_Binary.csv"
caminho_destino = "/Users/marcelosilva/Desktop/projectOne/6/B- Binary Target DS DT"

# Criar diretório de destino se não existir
os.makedirs(caminho_destino, exist_ok=True)

# Carregar o dataset binário
print("Carregando dataset binário...")
df = pd.read_csv(caminho_origem)

print(f"Dataset carregado com sucesso. Shape: {df.shape}")

# Verificar se a coluna target existe
if 'status_nutricional_who' not in df.columns:
    print("Erro: Coluna 'status_nutricional_who' não encontrada no dataset!")
    print("Colunas disponíveis:", df.columns.tolist())
else:
    # Mostrar distribuição da variável target
    print("\nDistribuição da variável target 'status_nutricional_who':")
    target_counts = df['status_nutricional_who'].value_counts().sort_index()
    print(target_counts)
    print(f"Proporção: {target_counts.values / len(df) * 100}")
    
    # Dividir em treino e teste (80/20) mantendo a proporção da variável target
    df_train, df_test = train_test_split(
        df, 
        test_size=0.2,           # 20% para teste
        random_state=42,         # Para reprodutibilidade
        stratify=df['status_nutricional_who']  # Manter proporção da variável target
    )
    
    print(f"\nDivisão realizada:")
    print(f"- Dataset de Treino: {len(df_train)} amostras ({len(df_train)/len(df)*100:.1f}%)")
    print(f"- Dataset de Teste: {len(df_test)} amostras ({len(df_test)/len(df)*100:.1f}%)")
    
    # Verificar se as proporções foram mantidas
    print("\nDistribuição no dataset de Treino:")
    train_counts = df_train['status_nutricional_who'].value_counts().sort_index()
    print(train_counts)
    print(f"Proporção: {train_counts.values / len(df_train) * 100}")
    
    print("\nDistribuição no dataset de Teste:")
    test_counts = df_test['status_nutricional_who'].value_counts().sort_index()
    print(test_counts)
    print(f"Proporção: {test_counts.values / len(df_test) * 100}")
    
    # Salvar os datasets
    arquivo_train = os.path.join(caminho_destino, "DatasetTrain.csv")
    arquivo_test = os.path.join(caminho_destino, "DatasetTest.csv")
    
    df_train.to_csv(arquivo_train, index=False)
    df_test.to_csv(arquivo_test, index=False)
    
    print(f"\nDatasets salvos com sucesso:")
    print(f"- Dataset de Treino: {arquivo_train}")
    print(f"- Dataset de Teste: {arquivo_test}")
    
    # Verificação de integridade
    print(f"\nVerificação de integridade:")
    print(f"- Total original: {len(df)} amostras")
    print(f"- Treino + Teste: {len(df_train) + len(df_test)} amostras")
    print(f"- Diferença: {len(df) - (len(df_train) + len(df_test))} (deve ser 0)")
    
    # Verificar se não há pacientes duplicados entre treino e teste
    # (assumindo que existe alguma coluna de ID ou usando o índice)
    indices_train = set(df_train.index)
    indices_test = set(df_test.index)
    intersecao = indices_train.intersection(indices_test)
    
    print(f"- Pacientes em comum entre treino e teste: {len(intersecao)} (deve ser 0)")
    
    if len(intersecao) == 0:
        print("✅ Divisão realizada com sucesso! Não há sobreposição entre os datasets.")
    else:
        print("❌ Erro: Há pacientes duplicados entre treino e teste!")
```
