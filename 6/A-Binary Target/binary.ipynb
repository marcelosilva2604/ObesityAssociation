```python
import pandas as pd
import os

# Definir caminhos
caminho_origem = "/Users/marcelosilva/Desktop/projectOne/4/F-TargetLabelEnconding/Test Dataset/DataSetTarget.csv"
caminho_destino = "/Users/marcelosilva/Desktop/projectOne/6/A-Binary Target"

# Criar diretório de destino se não existir
os.makedirs(caminho_destino, exist_ok=True)

# Carregar o dataset original
print("Carregando dataset original...")
df = pd.read_csv(caminho_origem)

# Verificar se a coluna 'status_nutricional_who' existe
if 'status_nutricional_who' not in df.columns:
    print("Erro: Coluna 'status_nutricional_who' não encontrada no dataset!")
    print("Colunas disponíveis:", df.columns.tolist())
else:
    print(f"Dataset carregado com sucesso. Shape: {df.shape}")
    
    # Mostrar distribuição original
    print("\nDistribuição original da variável 'status_nutricional_who':")
    print(df['status_nutricional_who'].value_counts().sort_index())
    
    # Fazer a transformação
    # 0 e 1 -> 0
    # 2 e 3 -> 1
    df_transformado = df.copy()
    df_transformado['status_nutricional_who'] = df['status_nutricional_who'].map({
        0: 0,
        1: 0,
        2: 1,
        3: 1
    })
    
    # Verificar se há valores não mapeados (NaN)
    valores_nao_mapeados = df_transformado['status_nutricional_who'].isna().sum()
    if valores_nao_mapeados > 0:
        print(f"\nAtenção: {valores_nao_mapeados} valores não foram mapeados (valores diferentes de 0, 1, 2, 3)")
        valores_originais_unicos = df['status_nutricional_who'].unique()
        print(f"Valores únicos originais: {sorted(valores_originais_unicos)}")
    
    # Mostrar distribuição transformada
    print("\nDistribuição após transformação:")
    print(df_transformado['status_nutricional_who'].value_counts().sort_index())
    
    # Salvar o dataset transformado
    arquivo_destino = os.path.join(caminho_destino, "DataSetTarget_Binary.csv")
    df_transformado.to_csv(arquivo_destino, index=False)
    
    print(f"\nDataset transformado salvo em: {arquivo_destino}")
    print(f"Shape do dataset final: {df_transformado.shape}")
    
    # Mostrar estatísticas finais
    print("\nResumo da transformação:")
    print(f"- Valores 0 e 1 (originais) -> 0 (novo): {(df['status_nutricional_who'].isin([0, 1])).sum()} registros")
    print(f"- Valores 2 e 3 (originais) -> 1 (novo): {(df['status_nutricional_who'].isin([2, 3])).sum()} registros")
    print(f"- Total de registros processados: {len(df)}")
```
